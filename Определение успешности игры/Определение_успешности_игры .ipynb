# Определение успешности игры <a id="head"></a>

Интернет-магазин «Стримчик» продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные до 2016 года о продажах игр, оценки пользователей и экспертов, жанры и платформы.

**Задача:** выявить определяющие успешность игры закономерности, что позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

**Описание данных:**
* `Name` — название игры
* `Platform` — платформа
* `Year_of_Release` — год выпуска
* `Genre` — жанр игры
* `NA_sales` — продажи в Северной Америке (миллионы проданных копий)
* `EU_sales` — продажи в Европе (миллионы проданных копий)
* `JP_sales` — продажи в Японии (миллионы проданных копий)
* `Other_sales` — продажи в других странах (миллионы проданных копий)
* `Critic_Score` — оценка критиков (максимум 100)
* `User_Score` — оценка пользователей (максимум 10)
* `Rating` — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

## Исследовательский анализ данных

### Сколько игр выпускалось в разные годы

Проверим колличество игр выпустившихся в разные годы, сгпуппировав данные по колонке `year_of_release`.

```python
df.groupby('year_of_release')['year_of_release'].count()
```

```python
year_count = df[df.duplicated(['name', 'year_of_release'], keep='first')]
year_count = year_count.groupby('year_of_release')['year_of_release'].count()
year_count
```

```python
year_count.plot(x='year_of_release', kind= 'bar', figsize = (15,5))
plt.xlabel('Год выпуска')
plt.ylabel('Количество игр')
plt.title('Выпуск игр');
```

Видим что пик выхода игр приходится на 2002 год, посчти в 4 раза больше чем год ранее. В открытых источников говорят что есть консоли нескольких поколений, и была большая конкуренция, игры становились лучше и увеличивалось количество жжелающих приобрести новые игры. в 2012 году видим спад выпука новых игр, возможно это было связано с мировым экономическим кризисом 2012-2013 гг. Так же возможен спад с появлением примерно с 2010" года смартфонов с быстрыми видеочипами и большим количеством оперативной памяти - которые оттянули на себя программирование и выпуск игр под игровые платформы.

```python
df.groupby('year_of_release')['year_of_release'].count().plot(x='year_of_release', kind= 'bar', figsize = (15,5))
plt.title('Количество выпущенных игр по годам')
plt.xlabel('Год выпуска')
plt.ylabel('Количество игр');
```

В связи с развитием индрустии и увеличением выпуска игр в 1994г (в 2 два раза, отноительно 1994г), можем предположить что данные за период до 1994г не существенно важны для анализа, а с 2013 более актуальными.

### Изменение продаж по платформам

Посмотрим сумму продаж по платформам от большего к меньшему, сгпуппировав данные по колонке `platform`.

```python
platforms = df.groupby('platform')['sum_sales'].sum().sort_values(ascending=False)
platforms
```

Для постройки распределений по годам платформ с наибольшими суммарными продажами копий, выберем первые 6 платформ с наибольшими значениями и создадим цикл для построение графиков.

```python
platforms = platforms.head(6).index

for p in platforms:
    (df
        .query('platform == @p')
        .pivot_table(index='year_of_release', values='sum_sales', aggfunc='sum')
        .plot(kind='bar', figsize=(12,4))
    )
    plt.title(f'Суммарные продажи по годам платформы {p}')
    plt.xlabel('Год выпуска')
    plt.ylabel('Продажи');
```

Пратформы PS2, DS, PS ушли из массового пользования, только X360, Wii, PS3 захватили 2016г. В среднем проходит 10 лет с появления до изчезания платформы. Видим графики с нормальным распределением, и у всех платформ с максимальными суммами количества продаж пики прошли.

### Актуальные данные, для постройки прогноза на 2017 год

В результате исследования предыдущих вопросов выбрали период с 2013 по 2016гг.

```python
df_current = df.query('year_of_release >= 2013')
```

#### Потенциально прибыльные платформы

Найдём потенциально прибыльные платформы по актуальному периоду

```python
platforms = df_current.groupby('platform')['sum_sales'].sum().sort_values(ascending=False)
platforms
```

```python
platforms = platforms.index
(df_current
     .query('platform in @platforms')
     .pivot_table(index='year_of_release', columns='platform', values='sum_sales', aggfunc='sum')
     .plot(figsize=(12,6))
)
plt.title('Суммарные продажи по годам')
plt.xlabel('Год выпуска')
plt.ylabel('Продажи');
```

Видим 2 потенциальные платформы PS4 и XOne. Платформа PS3 по сумме количества продаж стоит на 2-ом месте, но по графику продаж видно, что основной пик был в 2013 году и продажи игр в данной платформе падают. Помним, что данные за 2016 год могут быть неполными.

#### Диаграммы размаха

Построим график «ящик с усами» по глобальным продажам игр в разбивке по потенциально прибыльным платформам и узнаем числовое описание данных.

## Проверка гипотез

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Для сравнения данных о средних пользовательских рейтинах сформулируем нулевую и альтернативную гипотезы.
Нулевая гипотеза H0: средние пользовательские рейтинги платформ Xbox One и PC одинаковые.
Альтернативная гипотеза H1: средние пользовательские рейтинги платформ Xbox One и PC разные.

```python
# возьмём актуальный датасет без пропущенных значений в оценке пользователей и построим гистограмму распределения
user_score = df_current.dropna(subset=['user_score'])
user_score['user_score'].hist(bins=20);
```

Распределение пользовательских рейтингов склонно к нормальному, имеется скос в сторону меньших значений.

```python
# Применим t-тест для выборок по каждой платформе. Тк указанные выборки независимы equal_var = False. 
results = st.ttest_ind(user_score[user_score['platform'] == 'XOne']['user_score'],\
                       user_score[user_score['platform'] == 'PC']['user_score'], equal_var = False)

print('p-значение: ', results.pvalue)
```

Зададим alpha = .05 и сравним с p-значением.

```python
alpha = .05
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Полученное p-значение говорит о том, что вероятность получить случайно найденное различие между средними пользовательскими рейтингами платформ Xbox One и PC и при условии, что нулевая гипотеза верна, велика.
Выбранный критерий значимости был превышен. Следовательно, не получилось отвергнуть нулевую гипотезу.

### Средние пользовательские рейтинги жанров Action и Sports разные.

Для сравнения данных о средних пользовательских рейтинах жанров Action и Sports сформулирую нулевую и альтернативную гипотезы. Нулевая гипотеза H0: cредние пользовательские рейтинги жанров Action и Sports одинаковые. Альтернативная гипотеза H1: Cредние пользовательские рейтинги жанров Action и Sports разные.

```python
#Применю t-тест для выборок по каждому жанру. Тк указанные выборки независимы equal_var = False
#Для теста использую актуальный датасет без пропущенных значений в оценке пользователей
results = st.ttest_ind(user_score[user_score['genre'] == 'Action']['user_score'],\
                       user_score[user_score['genre'] == 'Sports']['user_score'], equal_var = False)
print('p-значение: ', results.pvalue)
```

Зададим alpha = .05 и сравним с p-значением.

```python
alpha = .05
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Полученное p-значение говорит о том, что вероятность получить случайно найденное различие между средними пользовательскими рейтингами жанров Action и Sports при условии, что нулевая гипотеза верна, стремиться к нулю. Выбранный критерий значимости не был превышен. Следовательно, рейтинги разные.
